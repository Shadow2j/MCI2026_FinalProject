# Analisis Sentimen NLP
Pastikan Anda telah mengunggah `order_reviews.csv`.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
import re
from collections import Counter

nltk.download('stopwords')
sns.set_theme(style="whitegrid")
```

## Memuat Data Ulasan

```
df_reviews = pd.read_csv('order_reviews.csv')
df_reviews = df_reviews.dropna(subset=['review_comment_message'])
df_reviews.head()
```

## Membersihkan Teks

```
stop_words = set(stopwords.words('portuguese'))
def clean_text(text):
    if not isinstance(text, str): return ""
    text = re.sub(r'[^a-záéíóúâêôãõç\s]', '', text.lower())
    return " ".join([w for w in text.split() if w not in stop_words and len(w)>2])

df_reviews['cleaned_text'] = df_reviews['review_comment_message'].apply(clean_text)
```

## Analisis Kata Teratas (Bintang 1)

```
bad_reviews = df_reviews[df_reviews['review_score'] == 1]['cleaned_text']
all_words = []
for t in bad_reviews: all_words.extend(t.split())
word_freq = Counter(all_words).most_common(20)
df_top = pd.DataFrame(word_freq, columns=['Kata', 'Jumlah'])
sns.barplot(x='Jumlah', y='Kata', data=df_top)
plt.title('Top Kata di Ulasan Buruk')
plt.show()
```
