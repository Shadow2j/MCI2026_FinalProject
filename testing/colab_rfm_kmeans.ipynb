# Segmentasi Pelanggan RFM Menggunakan K-Means Clustering
Pastikan Anda telah mengunggah file `data_rfm_colab.csv` ke dalam Colab ini.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Mengambil Data RFM

```
df_rfm = pd.read_csv('data_rfm_colab.csv')
print(f"Berhasil memuat {len(df_rfm)} baris data.")
df_rfm.head()
```

## 2. Pra-pemrosesan Data (Standardization)

```
scaler = StandardScaler()
rfm_features = df_rfm[['recency', 'frequency', 'monetary']]
rfm_scaled = scaler.fit_transform(rfm_features)
df_scaled = pd.DataFrame(rfm_scaled, columns=['recency_scaled', 'frequency_scaled', 'monetary_scaled'])
df_scaled.head()
```

## 3. Mencari Jumlah Kluster Optimal (Elbow Method)

```
sse = {}
for k in range(1, 9):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    sse[k] = kmeans.inertia_

plt.title('Metode Elbow')
sns.lineplot(x=list(sse.keys()), y=list(sse.values()), marker='o')
plt.show()
```

## 4. Melatih Model K-Means (K = 4)

```
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(df_scaled)
df_rfm['cluster'] = kmeans.labels_
df_rfm.head()
```

## 5. Analisis Profil Kluster

```
cluster_profile = df_rfm.groupby('cluster').agg({
    'recency': 'mean',
    'frequency': 'mean',
    'monetary': ['mean', 'count']
}).reset_index()
print(cluster_profile)
```
